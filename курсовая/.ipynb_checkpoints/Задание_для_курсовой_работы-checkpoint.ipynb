```python
#Task 1

print("–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
t=int(input())
assert 1<= t <= 1000, "–Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ t"
for j in range(t):
    n=int(input())
    assert 1<= n <= 2*105, "–Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ n"
    mas=list(input())
    x=0
    y=0
    #—Å–ª–æ–≤–∞—Ä—å –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç
    d={}
    d[(x,y)]=0
    l=-1
    r=n
    for i in range(n):
        if mas[i]=='R':
            y+=1
        elif mas[i]=='L':
            y-=1
        elif mas[i]=='U':
            x+=1
        elif mas[i]=='D':
            x-=1
        #
        if (x,y) in d:
            if i - d[(x,y)]<r-l:
                r=i
                l=d[(x,y)]
        d[(x,y)]=i+1
    print("–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
    if l==-1:
        print(-1)
    else:
        print(l+1,r+1)
```

```python
#task 2

print("–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
t = int(input())
assert 1<= t <= 1000, "–Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ t"
for j in range(t):
    n0, n1, n2 = map(int, input().split())
    assert 0<=n0<=100 and 0<=n1<=100 and 0<=n2<=100 and n0+n1+n2>0, "–Ω–µ–≤–µ—Ä–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ n"
    #–µ—Å–ª–∏ –µ–¥–∏–Ω–∏—Ü –Ω–µ—Ç, –∑–∞–ø–æ–ª–Ω—è–µ–º –Ω—É–ª—è–º–∏
    if n1 == 0 and n2 == 0:
        s = ('0' * (n0 + 1))
    #–µ—Å–ª–∏ –Ω—É–ª–µ–π –Ω–µ—Ç, –∑–∞–ø–æ–ª–Ω—è–µ–º –µ–¥–∏–Ω–∏—Ü–∞–º–∏
    elif n1 == 0 and n0 == 0:
        s = ('1' * (n2 + 1))
    else:
        s = '1'
        #–∑–∞–ø–æ–ª–Ω—è–µ–º –º–∞—Å—Å–∏–≤ –ø–µ—Ä–µ—Ö–æ–¥–∞–º–∏ –∏ –¥–æ–±–∞–≤–ª—è–µ–º –Ω–µ–æ–±—Ö –∫–æ–ª-–≤–æ 0 –∏–ª–∏ 1
        for i in range(n1):
            s += str(i % 2)
        s = s.replace('0', '0' * (n0 + 1), 1)
        s = s.replace('1', '1' * (n2 + 1), 1)
    print("–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
    print(s)
```

```python
#task 3

print("–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
n,m = map(int, input().split())
assert 2<=n<=2*105, "–Ω–µ–≤–µ—Ä–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ n"
assert 2<=m<=2*105, "–Ω–µ–≤–µ—Ä–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ m"
s = input()
assert len(s) == n, "–Ω–µ–≤–µ—Ä–Ω–∞—è –¥–ª–∏–Ω–∞ —Å—Ç—Ä–æ–∫–∏ s"
t = input()
assert len(t) == m, "–Ω–µ–≤–µ—Ä–Ω–∞—è –¥–ª–∏–Ω–∞ —Å—Ç—Ä–æ–∫–∏ t"

mas = []
i = 0
#–ø—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —Å–æ–≤–ø–∞–¥–µ–Ω–∏—è
for c in t:
    while s[i] != c:
        i += 1
    mas.append(i)
    i += 1

rs = []
i = n-1
#–ø—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —Å–æ–≤–ø–∞–¥–µ–Ω–∏—è —Ä–µ–≤–µ—Ä—Å
for c in t[::-1]:
    while s[i] != c:
        i -= 1
    rs.append(i)
    i -= 1
rs = rs[::-1]
print("–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
print(max(rs[i]-mas[i-1] for i in range(1,m)))
```

```python
#task 4

print("–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
n=int(input())
assert 1<=n<=100000, "–ù–µ–≤–µ—Ä–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ n"
#—Å–ª–æ–≤–∞—Ä—å —Å —ç–ª–µ–º–µ–Ω—Ç–∞–º–∏ - –º–Ω–æ–∂–µ—Å—Ç–≤–∞–º–∏
w={}
for i in range(n):
    w[i+1]=set()
mass={0}
r=0
q=list()
q.append((1,1,0))
for i in range(n-1):
    u,v=map(int,input().split())
    assert 1<=u<=n and 1<=v<=n and u!=v, "–Ω–µ–≤–µ—Ä–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è u –∏ v"
    w[u].add(v)
    w[v].add(u)
while q:
    u,v,c=q.pop()
    mass.add(u)
    o=0
    for i in w[u]:
        if i not in mass:
            o+=1
    if o:
        for i in w[u]:
            if i not in mass:
                q.append((i,v/o,c+1))
    else:r+=v*c
print("–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
print(r)
```

```python
#task 5

print("–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
t=int(input())
assert 1<=t<=104, "–Ω–µ–≤–µ—Ä–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ t"
for i in range(t):
    n=int(input())
    assert 2<=n<=105, "–Ω–µ–≤–µ—Ä–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ n"
    a=list(map(int, input().split()))
    assert len(a)==n, "–Ω–µ–≤–µ—Ä–Ω–∞—è –¥–ª–∏–Ω–∞ –º–∞—Å—Å–∏–≤–∞"
    b=[]
    c=[]
    for i in range(n-1):
        if a[i+1]==-1 and a[i]!=-1:
            b.append(a[i])
        elif a[i]==-1 and a[i+1]!=-1:
            b.append(a[i+1])
    if len(b)<=0:
        x=42
    else:
        x=(max(b)+min(b))//2
    for i in range(n):
        if a[i]==-1:
            a[i]=x
    for i in range(n-1):
        c.append(abs(a[i+1]-a[i]))
    print("–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
    print(max(c),x)
```

```python
#task 6

print("–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
n = int(input())
assert 2<=n<=2*105, "–Ω–µ–≤–µ—Ä–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ n"
a = list(map(int,input().split()))
assert len(a)==n, "–Ω–µ–≤–µ—Ä–Ω–∞—è –¥–ª–∏–Ω–∞ –º–∞—Å—Å–∏–≤–∞ –∞"
a1 = [1] * n
a2 = [1] * n
#–∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é –µ—Å–ª–∏ —Ç–µ–∫—É—â. –±–æ–ª—å—à–µ –ø—Ä–µ–¥—ã–¥—É—â.
for i in range (1, n):
    if a[i] > a[i - 1]:
        a1[i] = a1[i - 1] + 1
#–∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ —Å –∫–æ–Ω—Ü–∞ 
for i in range (n - 2, -1, -1):
    if a[i + 1] > a[i]:
        a2[i] = a2[i + 1] + 1
ans = max(a1)
#–ø—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å —É–¥–∞–ª–µ–Ω–∏—è —á–∏—Å–ª–∞
for i in range (n - 2):
    if a[i] < a[i + 2]:
        ans = max(ans, (a1[i] + a2[i + 2]))
print("–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
print(ans)
```

```python
#task 7

print("–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
n = int(input())
assert 2<=n<=200000, "–Ω–µ–≤–µ—Ä–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ n"
a = [int(x) for x in input().split()]
assert -109<=min(a) and max(a)<=109 and 0 not in a, "–Ω–µ–≤–µ—Ä–Ω–æ–µ –≤–≤–µ–¥–µ–Ω –º–∞—Å—Å–∏–≤ –∞"
d = set()
d.add(0)
cur = 0
ans = 0
#–ø–æ—Å–ª–µ–¥–æ–≤—Ç–µ–ª—å–Ω–æ —Å–∫–ª–∞–¥—ã–≤–∞–µ–º —á–∏—Å–ª–∞ –∏ –¥–æ–±–∞–≤–ª—è–µ–º –≤ –º–∞—Å—Å–∏–≤, –ø–æ–∫–∞ –Ω–µ –ø–æ—è–≤—è—Ç—Å—è –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ(0)
for i in range(n):
    cur += a[i]
    if cur in d:
        ans += 1
        d = set()
        d.add(0)
        cur = a[i]
    d.add(cur)
print("–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
print(ans)
```

```python
#task 8

import math
print("–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
n,a,b,k = map(int,input().split())
assert 1<=n<=2*105 and 1<=a<=109 and 1<=b<=109 and 1<=k<=109, "–Ω–µ–≤–µ—Ä–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –∏–≥—Ä—ã"
mas = list(map(int,input().split()))
assert 1<=min(mas) and max(mas)<=109, "–Ω–µ–≤–µ—Ä–Ω–æ –≤–≤–µ–¥–µ–Ω–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–¥–æ—Ä–æ–≤—å—è –º–æ–Ω—Å—Ç—Ä–æ–≤"
summ=0
points=0
list1 = list()
p = a+b
for i in range(len(mas)):
    if mas[i]%p == 0:
        list1.append(math.ceil(b/a))
    elif mas[i]%p >a:
        list1.append(math.ceil((mas[i]%p-a)/a))
    else:
        points+=1
list1.sort()
 
for i in range(len(list1)):
    summ+=list1[i]
    if summ>k:
        break
    points+=1
print("–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
print(points)
```

```python
#task 9

print("–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
t =int(input())
assert 1<=t<=1000, "–Ω–µ–≤–µ—Ä–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ t"
for _ in range(t):
    
    n = int(input())
    assert 1<=n<=50000, "–Ω–µ–≤–µ—Ä–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ t"
    ath = []
    for i in range(n):
        kol = list(map(int,input().split()))
        ath.append([int(i) for i in kol])
        assert 1<=min(kol) and max(kol)<=50000, "–Ω–µ–≤–µ—Ä–Ω–æ–µ –∑–Ω–∞—á–Ω–∏–µ –º–µ—Å—Ç–∞ —Å–ø–æ—Ä—Ç—Å–º–µ–Ω–∞"
    w = 0
 
    for i in range(1, n):
        count = 0
        for j in range(5):
            if ath[i][j] > ath[w][j]:
                count += 1
        if count < 3:
            w = i
    for i in range(n):
        if i == w:
            continue
        count = 0
        for j in range(5):
            if ath[i][j] > ath[w][j]:
                count += 1
        if count < 3:
            w = -2
            break
    print("–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ")
    print(w+1)
```
