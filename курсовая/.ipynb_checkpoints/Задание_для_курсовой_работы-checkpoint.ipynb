```python
#Task 1

print("ะฒัะพะดะฝัะต ะดะฐะฝะฝัะต")
t=int(input())
assert 1<= t <= 1000, "ะฝะตะดะพะฟัััะธะผะพะต ะทะฝะฐัะตะฝะธะต t"
for j in range(t):
    n=int(input())
    assert 1<= n <= 2*105, "ะฝะตะดะพะฟัััะธะผะพะต ะทะฝะฐัะตะฝะธะต n"
    mas=list(input())
    x=0
    y=0
    #ัะปะพะฒะฐัั ะดะปั ััะฐะฝะตะฝะธั ะบะพะพัะดะธะฝะฐั
    d={}
    d[(x,y)]=0
    l=-1
    r=n
    for i in range(n):
        if mas[i]=='R':
            y+=1
        elif mas[i]=='L':
            y-=1
        elif mas[i]=='U':
            x+=1
        elif mas[i]=='D':
            x-=1
        #
        if (x,y) in d:
            if i - d[(x,y)]<r-l:
                r=i
                l=d[(x,y)]
        d[(x,y)]=i+1
    print("ะฒััะพะดะฝัะต ะดะฐะฝะฝัะต")
    if l==-1:
        print(-1)
    else:
        print(l+1,r+1)
```

```python
#task 2

print("ะฒัะพะดะฝัะต ะดะฐะฝะฝัะต")
t = int(input())
assert 1<= t <= 1000, "ะฝะตะดะพะฟัััะธะผะพะต ะทะฝะฐัะตะฝะธะต t"
for j in range(t):
    n0, n1, n2 = map(int, input().split())
    assert 0<=n0<=100 and 0<=n1<=100 and 0<=n2<=100 and n0+n1+n2>0, "ะฝะตะฒะตัะฝะพะต ะทะฝะฐัะตะฝะธะต n"
    #ะตัะปะธ ะตะดะธะฝะธั ะฝะตั, ะทะฐะฟะพะปะฝัะตะผ ะฝัะปัะผะธ
    if n1 == 0 and n2 == 0:
        s = ('0' * (n0 + 1))
    #ะตัะปะธ ะฝัะปะตะน ะฝะตั, ะทะฐะฟะพะปะฝัะตะผ ะตะดะธะฝะธัะฐะผะธ
    elif n1 == 0 and n0 == 0:
        s = ('1' * (n2 + 1))
    else:
        s = '1'
        #ะทะฐะฟะพะปะฝัะตะผ ะผะฐััะธะฒ ะฟะตัะตัะพะดะฐะผะธ ะธ ะดะพะฑะฐะฒะปัะตะผ ะฝะตะพะฑั ะบะพะป-ะฒะพ 0 ะธะปะธ 1
        for i in range(n1):
            s += str(i % 2)
        s = s.replace('0', '0' * (n0 + 1), 1)
        s = s.replace('1', '1' * (n2 + 1), 1)
    print("ะฒััะพะดะฝัะต ะดะฐะฝะฝัะต")
    print(s)
```

```python
#task 3

print("ะฒัะพะดะฝัะต ะดะฐะฝะฝัะต")
n,m = map(int, input().split())
assert 2<=n<=2*105, "ะฝะตะฒะตัะฝะพะต ะทะฝะฐัะตะฝะธะต n"
assert 2<=m<=2*105, "ะฝะตะฒะตัะฝะพะต ะทะฝะฐัะตะฝะธะต m"
s = input()
assert len(s) == n, "ะฝะตะฒะตัะฝะฐั ะดะปะธะฝะฐ ัััะพะบะธ s"
t = input()
assert len(t) == m, "ะฝะตะฒะตัะฝะฐั ะดะปะธะฝะฐ ัััะพะบะธ t"

mas = []
i = 0
#ะฟัะพะฒะตัะบะฐ ะฝะฐ ัะพะฒะฟะฐะดะตะฝะธั
for c in t:
    while s[i] != c:
        i += 1
    mas.append(i)
    i += 1

rs = []
i = n-1
#ะฟัะพะฒะตัะบะฐ ะฝะฐ ัะพะฒะฟะฐะดะตะฝะธั ัะตะฒะตัั
for c in t[::-1]:
    while s[i] != c:
        i -= 1
    rs.append(i)
    i -= 1
rs = rs[::-1]
print("ะฒััะพะดะฝัะต ะดะฐะฝะฝัะต")
print(max(rs[i]-mas[i-1] for i in range(1,m)))
```

```python
#task 4

print("ะฒัะพะดะฝัะต ะดะฐะฝะฝัะต")
n=int(input())
assert 1<=n<=100000, "ะะตะฒะตัะฝะพะต ะทะฝะฐัะตะฝะธะต n"
#ัะปะพะฒะฐัั ั ัะปะตะผะตะฝัะฐะผะธ - ะผะฝะพะถะตััะฒะฐะผะธ
w={}
for i in range(n):
    w[i+1]=set()
mass={0}
r=0
q=list()
q.append((1,1,0))
for i in range(n-1):
    u,v=map(int,input().split())
    assert 1<=u<=n and 1<=v<=n and u!=v, "ะฝะตะฒะตัะฝัะต ะทะฝะฐัะตะฝะธั u ะธ v"
    w[u].add(v)
    w[v].add(u)
while q:
    u,v,c=q.pop()
    mass.add(u)
    o=0
    for i in w[u]:
        if i not in mass:
            o+=1
    if o:
        for i in w[u]:
            if i not in mass:
                q.append((i,v/o,c+1))
    else:r+=v*c
print("ะฒััะพะดะฝัะต ะดะฐะฝะฝัะต")
print(r)
```

```python
#task 5

print("ะฒัะพะดะฝัะต ะดะฐะฝะฝัะต")
t=int(input())
assert 1<=t<=104, "ะฝะตะฒะตัะฝะพะต ะทะฝะฐัะตะฝะธะต t"
for i in range(t):
    n=int(input())
    assert 2<=n<=105, "ะฝะตะฒะตัะฝะพะต ะทะฝะฐัะตะฝะธะต n"
    a=list(map(int, input().split()))
    assert len(a)==n, "ะฝะตะฒะตัะฝะฐั ะดะปะธะฝะฐ ะผะฐััะธะฒะฐ"
    b=[]
    c=[]
    for i in range(n-1):
        if a[i+1]==-1 and a[i]!=-1:
            b.append(a[i])
        elif a[i]==-1 and a[i+1]!=-1:
            b.append(a[i+1])
    if len(b)<=0:
        x=42
    else:
        x=(max(b)+min(b))//2
    for i in range(n):
        if a[i]==-1:
            a[i]=x
    for i in range(n-1):
        c.append(abs(a[i+1]-a[i]))
    print("ะฒััะพะดะฝัะต ะดะฐะฝะฝัะต")
    print(max(c),x)
```

```python
#task 6

print("ะฒัะพะดะฝัะต ะดะฐะฝะฝัะต")
n = int(input())
assert 2<=n<=2*105, "ะฝะตะฒะตัะฝะพะต ะทะฝะฐัะตะฝะธะต n"
a = list(map(int,input().split()))
assert len(a)==n, "ะฝะตะฒะตัะฝะฐั ะดะปะธะฝะฐ ะผะฐััะธะฒะฐ ะฐ"
a1 = [1] * n
a2 = [1] * n
#ะทะฐะฟะพะปะฝะตะฝะธะต ะผะฐััะธะฒะฐ ะฟะพ ะฒะพะทัะฐััะฐะฝะธั ะตัะปะธ ัะตะบัั. ะฑะพะปััะต ะฟัะตะดัะดัั.
for i in range (1, n):
    if a[i] > a[i - 1]:
        a1[i] = a1[i - 1] + 1
#ะทะฐะฟะพะปะฝะตะฝะธะต ะผะฐััะธะฒะฐ ั ะบะพะฝัะฐ 
for i in range (n - 2, -1, -1):
    if a[i + 1] > a[i]:
        a2[i] = a2[i + 1] + 1
ans = max(a1)
#ะฟัะพะฒะตัะบะฐ ะฝะฐ ะฒะพะทะผะพะถะฝะพััั ัะดะฐะปะตะฝะธั ัะธัะปะฐ
for i in range (n - 2):
    if a[i] < a[i + 2]:
        ans = max(ans, (a1[i] + a2[i + 2]))
print("ะฒััะพะดะฝัะต ะดะฐะฝะฝัะต")
print(ans)
```

```python
#task 7

print("ะฒัะพะดะฝัะต ะดะฐะฝะฝัะต")
n = int(input())
assert 2<=n<=200000, "ะฝะตะฒะตัะฝะพะต ะทะฝะฐัะตะฝะธะต n"
a = [int(x) for x in input().split()]
assert -109<=min(a) and max(a)<=109 and 0 not in a, "ะฝะตะฒะตัะฝะพะต ะฒะฒะตะดะตะฝ ะผะฐััะธะฒ ะฐ"
d = set()
d.add(0)
cur = 0
ans = 0
#ะฟะพัะปะตะดะพะฒัะตะปัะฝะพ ัะบะปะฐะดัะฒะฐะตะผ ัะธัะปะฐ ะธ ะดะพะฑะฐะฒะปัะตะผ ะฒ ะผะฐััะธะฒ, ะฟะพะบะฐ ะฝะต ะฟะพัะฒัััั ะพะดะธะฝะฐะบะพะฒัะต(0)
for i in range(n):
    cur += a[i]
    if cur in d:
        ans += 1
        d = set()
        d.add(0)
        cur = a[i]
    d.add(cur)
print("ะฒััะพะดะฝัะต ะดะฐะฝะฝัะต")
print(ans)
```

```python
#task 8

import math
print("ะฒัะพะดะฝัะต ะดะฐะฝะฝัะต")
n,a,b,k = map(int,input().split())
assert 1<=n<=2*105 and 1<=a<=109 and 1<=b<=109 and 1<=k<=109, "ะฝะตะฒะตัะฝัะต ะฟะฐัะฐะผะตััั ะธะณัั"
mas = list(map(int,input().split()))
assert 1<=min(mas) and max(mas)<=109, "ะฝะตะฒะตัะฝะพ ะฒะฒะตะดะตะฝะพ ะบะพะปะธัะตััะฒะพ ะทะดะพัะพะฒัั ะผะพะฝัััะพะฒ"
summ=0
points=0
list1 = list()
p = a+b
for i in range(len(mas)):
    if mas[i]%p == 0:
        list1.append(math.ceil(b/a))
    elif mas[i]%p >a:
        list1.append(math.ceil((mas[i]%p-a)/a))
    else:
        points+=1
list1.sort()
 
for i in range(len(list1)):
    summ+=list1[i]
    if summ>k:
        break
    points+=1
print("ะฒััะพะดะฝัะต ะดะฐะฝะฝัะต")
print(points)
```

```python
#task 9

print("ะฒัะพะดะฝัะต ะดะฐะฝะฝัะต")
t =int(input())
assert 1<=t<=1000, "ะฝะตะฒะตัะฝะพะต ะทะฝะฐัะตะฝะธะต t"
for _ in range(t):
    
    n = int(input())
    assert 1<=n<=50000, "ะฝะตะฒะตัะฝะพะต ะทะฝะฐัะตะฝะธะต t"
    ath = []
    for i in range(n):
        kol = list(map(int,input().split()))
        ath.append([int(i) for i in kol])
        assert 1<=min(kol) and max(kol)<=50000, "ะฝะตะฒะตัะฝะพะต ะทะฝะฐัะฝะธะต ะผะตััะฐ ัะฟะพัััะผะตะฝะฐ"
    w = 0
 
    for i in range(1, n):
        count = 0
        for j in range(5):
            if ath[i][j] > ath[w][j]:
                count += 1
        if count < 3:
            w = i
    for i in range(n):
        if i == w:
            continue
        count = 0
        for j in range(5):
            if ath[i][j] > ath[w][j]:
                count += 1
        if count < 3:
            w = -2
            break
    print("ะฒััะพะดะฝัะต ะดะฐะฝะฝัะต")
    print(w+1)
```
